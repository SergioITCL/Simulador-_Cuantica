```python

import quimb as qu
import quimb.tensor as qtn

# Definir las dimensiones de la red tensorial
Lx, Ly = 2, 2  # Tamaño de la red en dos dimensiones
D = 3          # Dimensión de los índices (puedes incrementarlo para mayor complejidad)

# Crear la red tensorial aleatoria
tn = qtn.TensorNetwork([])

# Rellenar la red con tensores aleatorios
for i in range(Lx):
    for j in range(Ly):
        # Crear etiquetas para los índices de los tensores, asegurando que sean únicos
        inds = (
            f'x{i},{j}_L',  # índice a la izquierda del tensor
            f'y{i},{j}_U',  # índice arriba del tensor
            f'x{i+1},{j}_R',  # índice a la derecha del tensor
            f'y{i},{j+1}_D'  # índice abajo del tensor
        )
        tensor = qu.randn((D, D, D, D), dtype='complex128')
        tn.add_tensor(qtn.Tensor(tensor, inds=inds))

# Mostrar la red tensorial
tn.draw()
```
